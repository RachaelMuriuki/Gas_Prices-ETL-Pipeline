```python
import os
import json
import pandas as pd
import http.client 
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv()
```

```python
#connect to API
conn = http.client.HTTPSConnection("api.collectapi.com")

headers = {
    'content-type': "application/json",
    'authorization': os.getenv("GAS_API_KEY")
    }

#request data
conn.request("GET", "/gasPrice/stateUsaPrice?state=WA", headers=headers)

#read response
res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```

```python
#convert from bytes to dict
data_string = data.decode("utf-8")
print(type(data_string))
data_dict = json.loads(data_string)
type(data_dict)
```

```python
#extract cities 
city_data = data_dict["result"]["cities"]

#convert to a dataframe
city_df = pd.DataFrame(city_data)   
```

```python
city_df.head()
```

```python
#drop lowername column
city_df = city_df.drop(columns=["lowername"])
city_df.head()
```

```python
#rename name to cities
city_df.rename(columns={"name": "cities"}, inplace=True)
city_df.head()
```
